```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create two random 3x3 matrices with integer values between 0 and 10

A = np.random.randint(0, 11, (3, 3))
B = np.random.randint(0, 11, (3, 3))

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
```

```python
# Perform matrix operations

addition = A + B
subtraction = A - B
multiplication = np.dot(A, B)

print("A+B",addition)
print("A-B",subtraction)
print("A*B",multiplication)
```

```python
# Determinant
detA = np.linalg.det(A)
print("Determinant of A:", detA)

# Rank
rankA = np.linalg.matrix_rank(A)
print("Rank", rankA)

# Inverse (if determinant â‰  0)

if detA != 0:
    inverseA = np.linalg.inv(A)
    print("Inverse of A:\n", inverseA)
else:
    print("Matrix A is singular. Inverse does not exist.")
```

```python
# Create two random 4x4 matrices with integer values between 0 and 10

X = np.random.randint(0, 11, (4, 4))
Y = np.random.randint(0, 11, (4, 4))

print("Matrix X:\n", X)
print("\nMatrix Y:\n", Y)
```

```python
add_xy = X + Y
sub_xy = X - Y
mul_xy = np.dot(X, Y)

print("\nX + Y:\n", add_xy)
print("\nX - Y:\n", sub_xy)
print("\nX x Y:\n", mul_xy)
```

```python
# Bar plot for Addition

plt.bar(range(16), add_xy.flatten())
plt.title("Bar Plot of X + Y")
plt.xlabel("Element Index")
plt.ylabel("Value")
plt.show()
```

```python
# Bar plot for Subtraction

plt.bar(range(16), sub_xy.flatten())
plt.title("Bar Plot of X - Y")
plt.xlabel("Element Index")
plt.ylabel("Value")
plt.show()
```

```python
# Bar plot for Multiplication

plt.bar(range(16), mul_xy.flatten())
plt.title("Bar Plot of X x Y")
plt.xlabel("Element Index")
plt.ylabel("Value")
plt.show()
```
